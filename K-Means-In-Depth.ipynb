```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns; sns.set() # For plot styling
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
# Generate a two-dimensional dataset containing four distinct blobs
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.show()
```

```python
# Apply the K-Means algorithm to the dataset
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
# Visualize the results using labels produced by K-Means
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.4)
plt.show()
```
